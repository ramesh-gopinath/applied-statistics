```python
# Electricity manufacturing company problem

# 2A. 

# Faiure rate = p = 5%
# so Success rate = q = 95%
# Sample - n = 10

#A) Probability of none of the items defective = p(0) = x = 0

# Since there are only 2 possibilities Success or Failure, we can use Binomial distribution to solve this problem

#P(x) = (n!/ (x! * (n-x))! * P^x * q^ (n-x)    
#P(0) = (10!/ (0! * (10-0))! * 0.05^0 * 0.95^(10-0)
#P(0) = (10! / 1 * 10!) * 1 * 0.95^10    =  10!/10! * 0.598737 = 0.598737

#So Probability of none of the items found defective = 0.598737 or 59.87%
```

```python
#A) Solving the 2A) question directly using python

import numpy as np
import pandas as pd
import scipy.stats as stats

rate = 0.05
k = np.arange(0,11)
n = 10
binom = stats.binom.pmf(k,n,rate)
print("Binom - Probability of none of the items found defective = ",binom[0])

# Probability of 0 items found defective is 59.87%
```

```python
# 2 B) Probability that exactly 1 of the item is defective p(1)

print("Probability of exactly 1 of the item is defective = ", binom[1])

# Probability of exactly 1 items found defective is 31.51%
```

```python
# 2 C) Probability that Two or fewer items will be found defective = p(x<=2) = p(0) + p(1) + p(2)

print(" Probability of 0 + 1 + 2 = ",binom[0] + binom[1] + binom[2])

# Probability of Two or fewer items found defective is - 98.85%
```

```python
# 2 D) Probability of Three or more items defective = p(x >= 3)

#= 1 - p(0) + p(1) + p(2)

print("Probability of three or more items found defective is = ",1 - (binom[0]+binom[1]+binom[2]))

# Probability of Three or more items found defective is 1.15%
```

```python
# 3 A. Car Salesman sells on an average 3 cars a week

rate1 = 3
n1 = np.arange(0,6)

poisson1 = stats.poisson.pmf(n1,rate1)

#A) Probability of selling some cars = p(x > 0) = or 1 - p(0)

print("Probability of selling some cars is ",1 - poisson1[0])

# Probability is 95.02%
```

```python
# 3 B Probability that he will sell 2 or more but lesser than 5 cars in a week

# p(2 >= x <5)

# so it will be poisson1[2] + poisson1[3] + poisson1[4] 

print("Probability of selling 2 or more but lesser than 5 cars is ", poisson1[2] + poisson1[3] + poisson1[4])

#Probability is 61.61%
```

```python
# 3 C Poisson distribution for cumulative probability of cars sold per week vs number of cars sold per week

import matplotlib.pyplot as plt

poisson1_cdf = stats.poisson.cdf(n1,rate1)

for i in poisson1, poisson1_cdf:
    print(i)
    
plt.plot(n1,poisson1_cdf,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate1)
plt.xlabel('Number of Cars sold per week')
plt.ylabel('Cumulative Probability of cars sold per week')
plt.show()
```

#4B. Write the Null hypothesis and alternate Hypothesis

Frequency of the initial population 

Operating    465
Closed       106
Acquired      86
Ipo            5

Frequency after dropping nulls 

Operating    319
Acquired      66
Closed        57
Ipo            4

Frequency after removing the outliers 

Operating    275
Closed        56
Acquired      55

Looking at the pattern of Operating startups vs Closed startups in the above 3 sample spaces, comparing the proportions of the 
companies and given the fact that Funding plays a very important factor in the longevity and continuation of the startups, we can define

Null hypothesis as : Distributions and proportions of Operating and closed companies will remain same

Alternate hypothesis : Distrbutions and proportions will not be the same and will change

Since we are working with a Categorical variable "Operating State" of the companies, we can use the Chi square test to solve this problem

Significance level : 95% confidence interval 

Data to compare

                          Observed data      Proportions        Expected        (O - E)           (O - E)^2 / E
Samples            Operating Closed Total Operating  Closed  Operating Closed Operating Closed Operating Closed

                     (A)      (B)    (C)    (D)       (E)      (F)      (G)     (H)      (I)      (J)      (K)
1  Original Data     465      106    571   81.43%    18.56%   473.15   97.84   -8.15     8.15     0.14     0.67      
2  minus nulls       319       57    376   84.84%    15.15%   311.56   64.43    7.43    -7.43     0.17     0.85
3  minus outliers    275       56    331   83.08%    16.91%   274.27   56.72    0.72    -0.72     0.00     0.00

4. Totals           1059      219   1278                      1059     219                        0.31     1.54

Calculations for Expected data

F1 = C1/C4 (Original data)/Totals * A4 
F2 = C2/C4 (Original data)/Totals * A4
F3 = C3/C4 (Original data)/Totals * A4

G1 = C1/C4 (Original data)/Totals * B4
G2 = C2/C4 (Original data)/Totals * B4
G3 = C3/C4 (Original data)/Totals * B4

Calculations for O - E

H1 = A1 - F1 
H2 = A2 - F2 
H3 = A3 - F3 

I1 = B1 - G1
I2 = B2 - G2
I3 = B3 - G3

Calculations for (O - E)^2 / E is self explained

Finally the Totals of J4 + K4 = 0.31 + 1.54 = 1.86 is the Chi square

Degress of freedom is (Rows - 1) * (columns - 1) = (3 -1) * (2 -1) = 2 * 1 = 2

So looking up the Chi square table for degress of freedom 2 and Chi square of 1.86 gives the following p values

chi square value of 1.386 is p value of 0.50 and
chi square value of 2.77 is p value of 0.25  

since our calculated chi square value 1.86 falls between 1.386 and 2.77 the 
p value is lying between 0.5 & 0.25 


Conclusion:

In our example, the Chi square 1.86 and degrees of freedom of 2 are associated with a P value of less than 0.50, but greater than 0.25. This means that a chi-square value this large or larger (or differences between expected and observed numbers this great or greater) would occur simply by chance between 25% and 50% of the time.  

Since the chi-square calculated value is less than the chi-square critical value, then we "fail to reject" our null hypothesis. 

Hence we infer proportions will not change.

Frequency in Operating startups 

Contestants  332

Logically speaking Startup companies which keep contesting in events will continue to be in Operation and also has a good chance to become winners.  So participating in Events becomes very critical for startup companies if they have to be in sustenance and continue to be in operating mode.  And it is only the Operating companies which will be in a position to Contest and win 
events.

Null hypothesis as : The two variables Operating state and Results are Independent

Alternate hypothesis : They both are Dependent variables

Since we are working with a Categorical variables "Operating State" & "Result" of the companies, we can use the Chi square test to solve this problem

Significance level : 95% confidence interval

4. Totals          45      820     1127                         45       820                             0.017    0.201

F1 = C1/C4 (Original data)/Totals * A4 
F2 = C2/C4 (Original data)/Totals * A4

G1 = C1/C4 (Original data)/Totals * B4
G2 = C2/C4 (Original data)/Totals * B4


Calculations for O - E

H1 = A1 - F1 
H2 = A2 - F2 

I1 = B1 - G1
I2 = B2 - G2

Calculations for (O - E)^2 / E is self explained

Finally the Totals of J4 + K4 = 0.017 + 0.201 = 0.209 is the Chi square value

Degress of freedom is (Rows - 1) * (columns - 1) = (2 -1) * (2 -1) = 1 * 1 = 1

So looking up the Chi square table for degress of freedom 1 and Chi square of 0.209 gives the following p values

chi square value of 0.102 is p value of 0.75 and

since our calculated chi square value 0.209 falls between 0.102 and 0.455 the 
p value is lying between 0.75 & 0.50....  


Conclusion:

In our example, the Chi square 0.209 and degrees of freedom of 1 are associated with a P value of less than 0.75, but greater than 0.50. This means that a chi-square value this large or larger (or differences between expected and observed numbers this great or greater) would occur simply by chance between 50% and 75% of the time.  

Since the chi-square calculated value is less than the chi-square critical value, we "fail to reject" our null hypothesis.
